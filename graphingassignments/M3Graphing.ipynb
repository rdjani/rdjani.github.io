```python
import matplotlib.pyplot as plt
import matplotlib.dates as dts
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('ConstructionTimeSeriesDataV2.csv')
```

```python
df.sort_values(by=['Month-Year'])
df.head(20)
```

```python
fig,ax = plt.subplots()
x = df['Month-Year']
y = df['Total Construction']
ax.plot(x,y)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_xlabel('Dates (Month-Year)',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Total Construction',fontsize='14',fontname='Times New Roman')
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20])
ax.set_ylim(0,max(y)+10)

ax.axvline(x=80,linewidth=1,color='k',label='dip in construction start',linestyle='-')
ax.axvline(x=122,linewidth=1,color='k',label='dip in construction ends',linestyle='-')
ax.legend(loc=1)
ax.axvspan(80,122,ymin=0,ymax = 300,facecolor='k',alpha=0.1)
ax.text(102,40000,'Around the 2008 Reccession',fontsize=14,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')

""" Enhance figure object """
fig.set_size_inches(14,8)
fig.suptitle("Time Series Plot of Temperature",fontsize='18',fontname='Times New Roman')
```

```python
'''In the cell above, we are graphing the time-series data of construction from January 2002 through 2013. The part that
I am specifically calling out is the signiciant dip in total construction around the September 2008 time period, as that is
near the beginning of the 2008 Recession in the United States. Although September 2009 was the highest recorded unemployment
period at the time, this stretch from September 2008 through early 2012 shows the difficult time period for the construction
industries. Even past early 2012, it was still struggling, but it did have a small improvement year after year.'''
```
